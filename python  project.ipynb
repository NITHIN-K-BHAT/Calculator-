## Scientific Calculator Using Tkinter

```python
# import tkinter as tk               # short form tikinter as tk
from tkinter import *              # here star(*) means take all
import math as m                   # importing math library 
root=Tk()
root.title('Scientific Calculator') # above line describes this lines
# root.geometry("361x381+500+200")                 # geometry

e=Entry(root,width=30,borderwidth=10,bg="powder blue",justify="right",font=("ariel",15))   #entry
e.grid(row=0,column=0,columnspan=5,padx=10,pady=15,ipadx=10,ipady=15)
root.configure(bg="#000000")      # background colour of the window

# on button click effect
def on_Button1(e):
    btn_one['bg']="#969696"
def on_Button2(e):
    btn_two['bg']="#969696"
def on_Button3(e):
    btn_three['bg']="#969696"
def on_Button4(e):
    btn_four['bg']="#969696"
def on_Button5(e):
    btn_five['bg']="#969696"
def on_Button6(e):
    btn_six['bg']="#969696"
def on_Button7(e):
    btn_seven['bg']="#969696"
def on_Button8(e):
    btn_eight['bg']="#969696"
def on_Button9(e):
    btn_nine['bg']="#969696"
def on_Buttonzero(e):
    btn_zero['bg']="#969696"
def on_Buttondoublezero(e):
    btn_doublezero['bg']="#969696"
def on_Buttondot(e):
    btn_dot['bg']="#969696"
def on_Buttonfact(e):
    btn_fact['bg']="#c6d9c3"
def on_Buttoninv(e):
    btn_inv['bg']="#c6d9c3"
def on_Buttonpara1(e):
    btn_para1['bg']="#c6d9c3"
def on_Buttonpara2(e):
    btn_para2['bg']="#c6d9c3"
def on_Buttonsin(e):
    btn_sin['bg']="#c6d9c3"
def on_Buttoncos(e):
    btn_cos['bg']="#c6d9c3"
def on_Buttontan(e):
    btn_tan['bg']="#969696"
def on_Buttonlog(e):
    btn_log['bg']="#c6d9c3"
def on_Buttonlan(e):
    btn_lan['bg']="#c6d9c3"
def on_Buttonpi(e):
    btn_pi['bg']="#c6d9c3"   
def on_Buttonbackspace(e):
    btn_backspace['bg']="#fc5151"
def on_Buttonclear(e):
    btn_clear['bg']="#fc5151"
def on_Buttonadd(e):
    btn_add['bg']="#c6d9c3"
def on_Buttonsub(e):
    btn_sub['bg']="#c6d9c3"
def on_Buttondiv(e):
    btn_div['bg']="#c6d9c3"
def on_Buttonmul(e):
    btn_mul['bg']="#c6d9c3"
def on_Buttonequal(e):
    btn_equal['bg']="#fc5151"  
    
# off button click effect  
def off_Button1(e):
    btn_one['bg']="#000000"
def off_Button2(e):
    btn_two['bg']="#000000"
def off_Button3(e):
    btn_three['bg']="#000000"
def off_Button4(e):
    btn_four['bg']="#000000"
def off_Button5(e):
    btn_five['bg']="#000000"
def off_Button6(e):
    btn_six['bg']="#000000"
def off_Button7(e):
    btn_seven['bg']="#000000"
def off_Button8(e):
    btn_eight['bg']="#000000"
def off_Button9(e):
    btn_nine['bg']="#000000"
def off_Buttonzero(e):
    btn_zero['bg']="#000000"
def off_Buttondoublezero(e):
    btn_doublezero['bg']="#000000"
def off_Buttondot(e):
    btn_dot['bg']="#000000"
def off_Buttonadd(e):
    btn_add['bg']="#000000"
def off_Buttonsub(e):
    btn_sub['bg']="#000000"
def off_Buttondiv(e):
    btn_div['bg']="#000000"
def off_Buttonmul(e):
    btn_mul['bg']="#000000"
def off_Buttonequal(e):
    btn_equal['bg']="#000000"
def off_Buttonfact(e):
    btn_fact['bg']="#000000"
def off_Buttoninv(e):
    btn_inv['bg']="#000000"
def off_Buttonpara1(e):
    btn_para1['bg']="#000000"
def off_Buttonpara2(e):
    btn_para2['bg']="#000000"
def off_Buttonsin(e):
    btn_sin['bg']="#000000"
def off_Buttoncos(e):
    btn_cos['bg']="#000000"
def off_Buttontan(e):
    btn_tan['bg']="#000000"
def off_Buttonlog(e):
    btn_log['bg']="#000000"
def off_Buttonlan(e):
    btn_lan['bg']="#000000"
def off_Buttonpi(e):
    btn_pi['bg']="#000000"
def off_Buttonbackspace(e):
    btn_backspace['bg']="#000000"
def off_Buttonclear(e):
    btn_clear['bg']="#000000"
    
    
     

def click(to_print):
    old=e.get()
    e.delete(0, END)
    e.insert(0, old+to_print)
    return
    
def clear(): 
    e.delete(0, END)  
    return

def backspace():
    current=e.get()
    length=len(current)-1
    e.delete(length, END)
    
def evaluate():
    ans=e.get()
    ans=eval(ans)
    e.delete(0, END)   # just to make entry delete and insert final answer 
    e.insert(0, ans)
    
def sc(event):
    key=event.widget
    text=key['text']
    no=e.get()
    result=''
    if text=='tan':
        result=str(m.tan(float(no)))
    if text=='sin':
        result=str(m.sin(float(no)))
    if text=='cos':
        result=str(m.cos(float(no)))
    if text=='log':
        result=str(m.log10(float(no)))
    if text==' lan':
        result=str(m.log(float(no)))
    if text=='x!':
        result=str(m.factorial(float(no)))
    if text=='inv':
        result=str(1/(float(no)))
    if text=='å…€ ':
        if no=="":
            result=str(m.pi)
        else:
            result=str(float(no) * m.pi)
    
    e.delete(0, END)   # just to make entry delete and insert final answer 
    e.insert(0, result)
    
def Exit_Func():
    root.destroy()


# button fact    
btn_fact=Button(root,text="x!",padx=32,pady=10,borderwidth=0,bg="#000000",fg="#ffffff")
btn_fact.bind("<Button-1>",sc)   
btn_fact.grid(row=1,column=0)
btn_fact.bind('<Enter>',on_Buttonfact)
btn_fact.bind('<Leave>',off_Buttonfact)

# button inverse
btn_inv=Button(root,text="inv",padx=28,pady=10,command=click,borderwidth=0,bg="#000000",fg="#ffffff")
btn_inv.bind("<Button-1>",sc)
btn_inv.grid(row=1,column=1)
btn_inv.bind('<Enter>',on_Buttoninv)
btn_inv.bind('<Leave>',off_Buttoninv)

# button para1
btn_para1=Button(root,text="( ",padx=34,pady=10,command=lambda: click("("),bg="#000000",fg="#ffffff",borderwidth=0)
btn_para1.grid(row=1,column=2)
btn_para1.bind('<Enter>',on_Buttonpara1)
btn_para1.bind('<Leave>',off_Buttonpara1)

# button para2
btn_para2=Button(root,text="  ) ",padx=37,pady=10,command=lambda: click(")"),bg="#000000",fg="#ffffff",borderwidth=0)
btn_para2.grid(row=1,column=3)
btn_para2.bind('<Enter>',on_Buttonpara2)
btn_para2.bind('<Leave>',off_Buttonpara2)

# button exit
exit_button=Button(root,text="Exit",command=Exit_Func,bg="red", fg="White",font=('italic',10),borderwidth=0,cursor="hand2")
exit_button.grid(row=1,column=4)


# button sin
btn_sin=Button(root,text="sin",padx=28.5,pady=10,borderwidth=0,bg="#000000",fg="#ffffff")
btn_sin.bind("<Button-1>",sc)   
btn_sin.grid(row=2,column=0)
btn_sin.bind('<Enter>',on_Buttonsin)
btn_sin.bind('<Leave>',off_Buttonsin)

# button cos
btn_cos=Button(root,text="cos",padx=27,pady=10,borderwidth=0,bg="#000000",fg="#ffffff")
btn_cos.bind("<Button-1>",sc)
btn_cos.grid(row=2,column=1)
btn_cos.bind('<Enter>',on_Buttoncos)
btn_cos.bind('<Leave>',off_Buttoncos)

# button tan
btn_tan=Button(root,text="tan",padx=29,pady=10,borderwidth=0,bg="#000000",fg="#ffffff")
btn_tan.bind("<Button-1>",sc)   
btn_tan.grid(row=2,column=2)
btn_tan.bind('<Enter>',on_Buttontan)
btn_tan.bind('<Leave>',off_Buttontan)

# button log
btn_log=Button(root,text="log",padx=35,pady=10,borderwidth=0,bg="#000000",fg="#ffffff")
btn_log.bind("<Button-1>",sc)
btn_log.grid(row=2,column=3)
btn_log.bind('<Enter>',on_Buttonlog)
btn_log.bind('<Leave>',off_Buttonlog)


# button lan
btn_lan=Button(root,text=" lan",padx=29,pady=10,borderwidth=0,bg="#000000",fg="#ffffff")
btn_lan.bind("<Button-1>",sc)
btn_lan.grid(row=2,column=4)
btn_lan.bind('<Enter>',on_Buttonlan)
btn_lan.bind('<Leave>',off_Buttonlan)

# button seven 
btn_seven=Button(root,text="7",padx=33,pady=10,command=lambda:click("7"),borderwidth=0,bg="#000000",fg="#ffffff")
btn_seven.grid(row=3,column=0)
btn_seven.bind('<Enter>',on_Button7)
btn_seven.bind('<Leave>',off_Button7)

# button eight
btn_eight=Button(root,text="8",padx=32,pady=10,command=lambda:click("8"),borderwidth=0,bg="#000000",fg="#ffffff")
btn_eight.grid(row=3,column=1)
btn_eight.bind('<Enter>',on_Button8)
btn_eight.bind('<Leave>',off_Button8)

# button nine
btn_nine=Button(root,text="9",padx=32,pady=10,command=lambda:click("9"),borderwidth=0,bg="#000000",fg="#ffffff")
btn_nine.grid(row=3,column=2)
btn_nine.bind('<Enter>',on_Button9)
btn_nine.bind('<Leave>',off_Button9)

# button backspace
btn_backspace=Button(root,text="âŒ«",padx=32,pady=10,command=backspace,relief=FLAT,activebackground="#ff0000",bg="#000000",fg="#ffffff")
btn_backspace.grid(row=3,column=3)
btn_backspace.bind('<Enter>',on_Buttonbackspace)
btn_backspace.bind('<Leave>',off_Buttonbackspace)

# button clear
btn_clear=Button(root,text="C",padx=32,pady=10,command=clear,relief=FLAT,bg="#000000",activebackground="#ff0000",fg="#ffffff")
btn_clear.grid(row=3,column=4)
btn_clear.bind('<Enter>',on_Buttonclear)
btn_clear.bind('<Leave>',off_Buttonclear)

# button four
btn_four=Button(root,text="4",padx=33,pady=10,command=lambda:click("4"),borderwidth=0,bg="#000000",fg="#ffffff")
btn_four.grid(row=4,column=0)
btn_four.bind('<Enter>',on_Button4)
btn_four.bind('<Leave>',off_Button4)

# button five
btn_five=Button(root,text="5",padx=32,pady=10,command=lambda:click("5"),borderwidth=0,bg="#000000",fg="#ffffff")
btn_five.grid(row=4,column=1)
btn_five.bind('<Enter>',on_Button5)
btn_five.bind('<Leave>',off_Button5)

# button six
btn_six=Button(root,text="6",padx=32,pady=10,command=lambda:click("6"),borderwidth=0,bg="#000000",fg="#ffffff")
btn_six.grid(row=4,column=2)
btn_six.bind('<Enter>',on_Button6)
btn_six.bind('<Leave>',off_Button6)

# button mul
btn_mul=Button(root,text="*",padx=37,pady=10,command=lambda:click('*'),relief=FLAT,bg="#000000",fg="#ffffff")
btn_mul.grid(row=4,column=3)
btn_mul.bind('<Enter>',on_Buttonmul)
btn_mul.bind('<Leave>',off_Buttonmul)

# button div
btn_div=Button(root,text="/",padx=34,pady=10,command=lambda:click('/'),relief=FLAT,bg="#000000",fg="#ffffff")
btn_div.grid(row=4,column=4)
btn_div.bind('<Enter>',on_Buttondiv)
btn_div.bind('<Leave>',off_Buttondiv)

# button one
btn_one=Button(root,text="1",padx=33,pady=10,command=lambda:click("1"),borderwidth=0,bg="#000000",fg="#ffffff")
btn_one.grid(row=5,column=0)
btn_one.bind('<Enter>',on_Button1)
btn_one.bind('<Leave>',off_Button1)

# button two
btn_two=Button(root,text="2",padx=32,pady=10,command=lambda:click("2"),borderwidth=0,bg="#000000",fg="#ffffff")
btn_two.grid(row=5,column=1)
btn_two.bind('<Enter>',on_Button2)
btn_two.bind('<Leave>',off_Button2)

# button three
btn_three=Button(root,text="3",padx=32,pady=10,command=lambda:click("3"),borderwidth=0,bg="#000000",fg="#ffffff")
btn_three.grid(row=5,column=2)
btn_three.bind('<Enter>',on_Button3)
btn_three.bind('<Leave>',off_Button3)

# button add
btn_add=Button(root,text="+ ",padx=34,pady=10,command=lambda:click('+'),relief=FLAT,bg="#000000",fg="#ffffff")
btn_add.grid(row=5,column=3)
btn_add.bind('<Enter>',on_Buttonadd)
btn_add.bind('<Leave>',off_Buttonadd)

# button sub
btn_sub=Button(root,text="-",padx=34.2,pady=10,command=lambda:click('-'),relief=FLAT,bg="#000000",fg="#ffffff")
btn_sub.grid(row=5,column=4)
btn_sub.bind('<Enter>',on_Buttonsub)
btn_sub.bind('<Leave>',off_Buttonsub)

# button zero
btn_zero=Button(root,text="0",padx=32,pady=10,command=lambda:click("0"),borderwidth=0,bg="#000000",fg="#ffffff")
btn_zero.grid(row=6,column=0)
btn_zero.bind('<Enter>',on_Buttonzero)
btn_zero.bind('<Leave>',off_Buttonzero)

# button dot
btn_dot=Button(root,text=".",padx=33.5,pady=10,command=lambda:click("."),borderwidth=0,bg="#000000",fg="#ffffff")
btn_dot.grid(row=6,column=1)
btn_dot.bind('<Enter>',on_Buttondot)
btn_dot.bind('<Leave>',off_Buttondot)

# button doublezero
btn_doublezero=Button(root,text="00",padx=29,pady=10,command=lambda:click("00"),borderwidth=0,bg="#000000",fg="#ffffff")
btn_doublezero.grid(row=6,column=2)
btn_doublezero.bind('<Enter>',on_Buttondoublezero)
btn_doublezero.bind('<Leave>',off_Buttondoublezero)

# button pi
btn_pi= Button(root, text="å…€ ", padx=32, pady=10,relief=FLAT,bg="#000000",fg="#ffffff")
btn_pi.bind("<Button-1>", sc)
btn_pi.grid(row=6,column=3)
btn_pi.bind('<Enter>',on_Buttonpi)
btn_pi.bind('<Leave>',off_Buttonpi)

# button equal
btn_equal=Button(root,text="=",padx=32,pady=10,command=evaluate,relief=FLAT,bg="#000000",activebackground="#ff0000",fg="#ffffff")
btn_equal.grid(row=6,column=4)
btn_equal.bind('<Enter>',on_Buttonequal)
btn_equal.bind('<Leave>',off_Buttonequal)


    
 # run the application   
root.mainloop()      
```
